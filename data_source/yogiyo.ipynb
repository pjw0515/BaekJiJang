```python
import pandas as pd
import numpy as np
```

```python
dust201301 = pd.read_csv("../data/folder/dust201301.csv", encoding='euc-kr')
dust201302 = pd.read_csv("../data/folder/dust201302.csv", encoding='euc-kr')
dust201303 = pd.read_csv("../data/folder/dust201303.csv", encoding='euc-kr')
dust201304 = pd.read_csv("../data/folder/dust201304.csv", encoding='euc-kr')

dust201401 = pd.read_csv("../data/folder/dust201401.csv", encoding='euc-kr')
dust201402 = pd.read_csv("../data/folder/dust201402.csv", encoding='euc-kr')
dust201403 = pd.read_csv("../data/folder/dust201403.csv", encoding='euc-kr')
dust201404 = pd.read_csv("../data/folder/dust201404.csv", encoding='euc-kr')

dust201501 = pd.read_csv("../data/folder/dust201501.csv", encoding='utf-8')
dust201502 = pd.read_csv("../data/folder/dust201502.csv", encoding='utf-8')
dust201503 = pd.read_csv("../data/folder/dust201503.csv", encoding='utf-8')
dust201504 = pd.read_csv("../data/folder/dust201504.csv", encoding='utf-8')

dust201601 = pd.read_csv("../data/folder/dust201601.csv", encoding='euc-kr')
dust201602 = pd.read_csv("../data/folder/dust201602.csv", encoding='euc-kr')
dust201603 = pd.read_csv("../data/folder/dust201603.csv", encoding='euc-kr')
dust201604 = pd.read_csv("../data/folder/dust201604.csv", encoding='euc-kr')

dust201701 = pd.read_csv("../data/folder/dust201701.csv", encoding='euc-kr')
dust201702 = pd.read_csv("../data/folder/dust201702.csv", encoding='euc-kr')
dust201703 = pd.read_csv("../data/folder/dust201703.csv", encoding='euc-kr')
dust201704 = pd.read_csv("../data/folder/dust201704.csv", encoding='euc-kr')
```

```python
temp2013 = pd.concat([dust201301, dust201302])
temp2013 = pd.concat([temp2013, dust201303])
temp2013 = pd.concat([temp2013, dust201304])
temp2013['시간'] = temp2013['측정일시'].astype(str).str.slice(8,10,1)
temp2013['날짜'] = temp2013['측정일시'].astype(str).str.slice(0,8,1)

temp2014 = pd.concat([dust201401, dust201402])
temp2014 = pd.concat([temp2014, dust201403])
temp2014 = pd.concat([temp2014, dust201404])
temp2014['시간'] = temp2014['측정일시'].astype(str).str.slice(8,10,1)
temp2014['날짜'] = temp2014['측정일시'].astype(str).str.slice(0,8,1)

temp2015 = pd.concat([dust201501, dust201502])
temp2015 = pd.concat([temp2015, dust201503])
temp2015 = pd.concat([temp2015, dust201504])
temp2015['시간'] = temp2015['측정일시'].astype(str).str.slice(8,10,1)
temp2015['날짜'] = temp2015['측정일시'].astype(str).str.slice(0,8,1)

temp2016 = pd.concat([dust201601, dust201602])
temp2016 = pd.concat([temp2016, dust201603])
temp2016 = pd.concat([temp2016, dust201604])
temp2016['시간'] = temp2016['측정일시'].astype(str).str.slice(8,10,1)
temp2016['날짜'] = temp2016['측정일시'].astype(str).str.slice(0,8,1)

temp2017 = pd.concat([dust201701, dust201702])
temp2017 = pd.concat([temp2017, dust201703])
temp2017 = pd.concat([temp2017, dust201704])
temp2017['시간'] = temp2017['측정일시'].astype(str).str.slice(8,10,1)
temp2017['날짜'] = temp2017['측정일시'].astype(str).str.slice(0,8,1)
```

```python
east2013 = temp2013[temp2013['지역'] == '광주 동구']
east2014 = temp2014[temp2014['지역'] == '광주 동구']
east2015 = temp2015[temp2015['지역'] == '광주 동구']
east2016 = temp2016[temp2016['지역'] == '광주 동구']
east2017 = temp2017[temp2017['지역'] == '광주 동구']
```

```python
east2013 = east2013[east2013['시간'] == '14']
east2014 = east2014[east2014['시간'] == '14']
east2015 = east2015[east2015['시간'] == '14']
east2016 = east2016[east2016['시간'] == '14']
east2017 = east2017[east2017['시간'] == '14']
```

```python
east2013.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
east2014.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
east2015.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
east2016.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
east2017.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
```
