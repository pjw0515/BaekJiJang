```python
import pandas as pd
import numpy as np
```

```python
dust201601 = pd.read_csv("../data/folder/dust201601.csv", encoding='euc-kr')
dust201602 = pd.read_csv("../data/folder/dust201602.csv", encoding='euc-kr')
dust201603 = pd.read_csv("../data/folder/dust201603.csv", encoding='euc-kr')
dust201604 = pd.read_csv("../data/folder/dust201604.csv", encoding='euc-kr')
```

```python
temp = pd.concat([dust201601, dust201602])
temp = pd.concat([temp, dust201603])
temp = pd.concat([temp, dust201604])
temp['시간'] = temp['측정일시'].astype(str).str.slice(8,10,1)
temp['날짜'] = temp['측정일시'].astype(str).str.slice(0,8,1)
temp.head()
```

```python
east = temp[temp['지역'] == '광주 동구']
east.head()
```

```python
east = east[east['시간'] == '14']
east.head()
```

```python
east.rename(columns = {'PM10': '미세먼지', 'PM25': '초미세먼지'}, inplace=True)
east.head()
```

```python
dates = pd.date_range('20160101', periods = 366)
dates
```

```python
east['측정날짜'] = dates
```

```python
east.head()
```

```python
del east['측정일시']
del east['날짜']
del east['시간']
```

```python
east.head()
```

```python
east.set_index(dates, inplace=True)
east.head()
```

```python
east['일시'] = dates
```

```python
east.head()
```

```python
chosun_weather = pd.read_csv("../data/chosun2016.csv", encoding = 'euc-kr')
chosun_weather.head()
```

```python
chosun_weather.set_index(dates, inplace=True)
chosun_weather.head()
```

```python
result = pd.concat([east, chosun_weather], axis=1, join_axes=[east.index])
result
```
