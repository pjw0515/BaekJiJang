```python
import pandas as pd
import numpy as np
```

```python
dust201601 = pd.read_csv("../data/folder/dust201601.csv", encoding='euc-kr')
dust201602 = pd.read_csv("../data/folder/dust201602.csv", encoding='euc-kr')
dust201603 = pd.read_csv("../data/folder/dust201603.csv", encoding='euc-kr')
dust201604 = pd.read_csv("../data/folder/dust201604.csv", encoding='euc-kr')
```

```python
temp = pd.concat([dust201601, dust201602])
```

```python
temp = pd.concat([temp, dust201603])
```

```python
temp = pd.concat([temp, dust201604])
```

```python
temp['시간'] = temp['측정일시'].astype(str).str.slice(8,10,1)
temp['날짜'] = temp['측정일시'].astype(str).str.slice(0,8,1)
temp.head()
```

```python
del temp['측정소코드']
del temp['SO2']
del temp['CO']
del temp['O3']
del temp['NO2']
del temp['주소']
del temp['측정일시']
```

```python
east = temp[temp['지역'] == '광주 동구']
west = temp[temp['지역'] == '광주 서구']
south = temp[temp['지역'] == '광주 남구']
north = temp[temp['지역'] == '광주 북구']
mid = temp[temp['지역'] == '광주 광산구']
west_nong = west[west['측정소명'] == '농성동']
west_chi = west[west['측정소명'] == '치평동']
north_dooam = north[north['측정소명'] == '두암동']
north_unnam = north[north['측정소명'] == '운암동']
north_kunkuk = north[north['측정소명'] == '건국동']
mid_songjeong = mid[mid['측정소명'] == '송정1동']
mid_oseon = mid[mid['측정소명'] == '오선동']
```

```python
east.head()
```

```python
east = east[east['시간'] == '14']
south = south[south['시간'] == '14']
west_nong = west_nong[west_nong['시간'] == '14']
west_chi = west_chi[west_chi['시간'] == '14']
north_dooam = north_dooam[north_dooam['시간'] == '14']
north_unnam = north_unnam[north_unnam['시간'] == '14']
north_kunkuk = north_kunkuk[north_kunkuk['시간'] == '14']
mid_songjeong = mid_songjeong[mid_songjeong['시간'] == '14']
mid_oseon = mid_oseon[mid_oseon['시간'] == '14']
```
