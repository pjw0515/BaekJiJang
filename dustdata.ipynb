```python
import pandas as pd
```

```python
dust = pd.read_csv('../2017/dust2017_1.csv', encoding='utf-8')
dust.head()
```

```python
del dust['측정소코드']
del dust['SO2']
del dust['CO']
del dust['O3']
del dust['NO2']

dust.head()
```

```python
dust.rename(columns = {'PM10': '미세먼지',
                      'PM25': '초미세먼지'}, inplace=True)
dust.head()
```

```python
dust.to_csv('../2017/dust1.csv', sep=',', encoding='euc-kr')
```
